```python
def get_hwb_class(row):
    # Major bridges
    if ((row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] < 1990) or 
        (row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] < 1975)) and row["MAJOR_BRIDGE"] == "Yes":
        return 1
    if ((row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] >= 1990) or 
        (row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] >= 1975)) and row["MAJOR_BRIDGE"] == "Yes":
        return 2
        
    # Single span bridges
    if ((row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] < 1990) or 
        (row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] < 1975)) and row["MAIN_UNIT_SPANS_045"] == 1:
        return 3
    if ((row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] >= 1990) or 
        (row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] >= 1975)) and row["MAIN_UNIT_SPANS_045"] == 1:
        return 4
        
    # Concrete bridges
    if row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] < 1990 and row["STRUCTURE_KIND_NAME"] == "Concrete":
        return 5
    if row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] < 1975 and row["STRUCTURE_KIND_NAME"] == "Concrete":
        return 6
    if ((row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] >= 1990) or 
        (row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] >= 1975)) and row["STRUCTURE_KIND_NAME"] == "Concrete":
        return 7
        
    # Concrete continuous bridges
    if row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] < 1975 and row["STRUCTURE_KIND_NAME"] == "Concrete continuous":
        return 8
    if row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] < 1975 and row["STRUCTURE_KIND_NAME"] == "Concrete continuous":
        return 9
    if ((row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] < 1990) or 
        (row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] < 1975)) and row["STRUCTURE_KIND_NAME"] == "Concrete continuous":
        return 10
    if ((row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] >= 1990) or 
        (row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] >= 1975)) and row["STRUCTURE_KIND_NAME"] == "Concrete continuous":
        return 11
        
    # Steel bridges
    if row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] < 1990 and row["STRUCTURE_KIND_NAME"] == "Steel":
        return 12
    if row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] < 1975 and row["STRUCTURE_KIND_NAME"] == "Steel":
        return 13
    if ((row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] >= 1990) or 
        (row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] >= 1975)) and row["STRUCTURE_KIND_NAME"] == "Steel":
        return 14
        
    # Steel continuous bridges
    if ((row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] < 1990) or 
        (row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] < 1975)) and row["STRUCTURE_KIND_NAME"] == "Steel continuous" and row["LENGTH_LESS_THAN_20"] == "No":
        return 15
    if ((row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] >= 1990) or 
        (row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] >= 1975)) and row["STRUCTURE_KIND_NAME"] == "Steel continuous":
        return 16
        
    # Prestressed concrete bridges
    if row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] < 1990 and row["STRUCTURE_KIND_NAME"] == "Prestressed concrete":
        return 17
    if row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] < 1975 and row["STRUCTURE_KIND_NAME"] == "Prestressed concrete":
        return 18
    if ((row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] >= 1990) or 
        (row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] >= 1975)) and row["STRUCTURE_KIND_NAME"] == "Prestressed concrete":
        return 19
        
    # Prestressed concrete continuous bridges
    if row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] < 1975 and row["STRUCTURE_KIND_NAME"] == "Prestressed concrete continuous":
        return 20
    if row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] >= 1975 and row["STRUCTURE_KIND_NAME"] == "Prestressed concrete continuous":
        return 21
        
    # Short steel bridges
    if row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] < 1990 and row["STRUCTURE_KIND_NAME"] == "Steel" and row["LENGTH_LESS_THAN_20"] == "Yes":
        return 24
    if row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] < 1975 and row["STRUCTURE_KIND_NAME"] == "Steel" and row["LENGTH_LESS_THAN_20"] == "Yes":
        return 25
    if row["STATE_CLASS"] == "NON-CA" and row["YEAR_BUILT_027"] < 1990 and row["STRUCTURE_KIND_NAME"] == "Steel continuous" and row["LENGTH_LESS_THAN_20"] == "Yes":
        return 26
    if row["STATE_CLASS"] == "CA" and row["YEAR_BUILT_027"] < 1975 and row["STRUCTURE_KIND_NAME"] == "Steel continuous" and row["LENGTH_LESS_THAN_20"] == "Yes":
        return 27
        
    # Default class
    return 28

df["HWB_CLASS"] = df.apply(get_hwb_class, axis=1)

df
```

```python
def k3d_val(row):
  k3d_class = row["K_3D_CLASS"]
  span = row["MAIN_UNIT_SPANS_045"]

  if k3d_class == 1:
      A, B = 0.25, 1.
  elif k3d_class == 2:
      A, B = 0.33, 0.
  elif k3d_class == 3:
      A, B = 0.33, 1.
  elif k3d_class == 4:
      A, B = 0.09, 1.
  elif k3d_class == 5:
      A, B = 0.05, 0.
  elif k3d_class == 6:
      A, B = 0.2, 1.
  elif k3d_class == 7:
      A, B = 0.1, 0.
  else:  
      A, B = 0., 100.

  res = span - B if span-B > 0.1 else 1.0
  return 1 + A/res

df["K_3D_VAL"] = df.apply(k3d_val, axis=1)
df
```

```python
def k_skew(row):
  skew = row["DEGREES_SKEW_034"]

  if skew > 90:
    skew = 69
  
  K_skew = np.sqrt(np.sin((90.-skew)/180.*np.pi))
  return K_skew

df["K_SKEW"] = df.apply(k_skew, axis=1)
df
```

```python
# convert latitude DMS format to decimal
def convert_decimal(row):
    lat = row["LAT_016"]
    
    if pd.isna(lat):
        return float('nan')
        
    degrees = int(lat / 1000000) 
    minutes = int((lat % 1000000) / 10000)
    seconds = (lat % 10000) / 100.0
    decimal_degrees = degrees + minutes / 60 + seconds / 3600
    return decimal_degrees

df["LATITUDE"] = df.apply(convert_decimal, axis=1)
df = df.drop(columns=["LAT_016"])
df
```

```python
def convert_decimal_lon(row):
    lon = row["LONG_017"]
    
    if pd.isna(lon):
        return float('nan')
        
    degrees = int(lon / 1000000) 
    minutes = int((lon % 1000000) / 10000)
    seconds = (lon % 10000) / 100.0
    decimal_degrees = degrees + minutes / 60 + seconds / 3600
    decimal_degrees *= -1
    return decimal_degrees

df["LONGITUDE"] = df.apply(convert_decimal_lon, axis=1)
df = df.drop(columns=["LONG_017"])
df
```

```python
cols = df.columns.tolist()
cols = ['LATITUDE', 'LONGITUDE'] + [col for col in cols if col not in ['LATITUDE', 'LONGITUDE']]
df = df[cols]
df
```
